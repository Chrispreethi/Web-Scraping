```python
#installing packages

import pandas as pd
from bs4 import BeautifulSoup
import requests
import re
import wget
import json
```

```python
#send request
main_url="https://en.wikipedia.org/w/index.php?title=Category:Male_actors_in_Malayalam_cinema"
main_page=requests.get(main_url)
soup=BeautifulSoup(main_page.text,'html.parser')
category=soup.find('div', class_='mw-category')
```

```python
#read a tag and capture href link
links = []
for link in category.find_all('a'):
    links.append(link.get('href'))
links

#list of links
root='https://en.wikipedia.org{}'
list_of_urls=[]
for i in links:
    list_of_urls.append(root.format(i))
    #print(list_of_urls)
    
list_of_urls
```

```python
#getting the tag"th"
th_tag=[]
for url in list_of_urls:
    # You can access the url in this loop with the name 'url'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    infobox = soup.find_all('table', class_='infobox biography vcard')
    for x in infobox: 
        tag1 = x('tr')
        for item in tag1:
            th_tag.append(item.getText().replace('\n','').strip())
th_tag = [sub.replace('\xa0', '') for sub in th_tag]
#th_tag = re.sub('([A-Z])', r'\1 ', str(th_tag))
th_tag
#tag
```

```python
abc =json.dumps(th_tag)
abc
```

```python
#getting the tag"th"
td_tag=[]
for url in list_of_urls:
    # You can access the url in this loop with the name 'url'
    page = requests.get(url)
    soup = BeautifulSoup(page.text, "html.parser")
    infobox = soup.find_all('table', class_='infobox biography vcard')
    for x in infobox: 
        tag1 = x('td') 
        for item in tag1:
            td_tag.append(item.getText().replace('\n',''))
td_tag = [sub.replace('\xa0', '') for sub in td_tag]
td_tag
```

```python
#create dataframe
final_array=[]
for var1,var2 in zip(th_tag,td_tag):
    final_array.append({'Variable1':var1,'Variable2':var2})
df=pd.DataFrame(final_array)
df.head(40)
```

```python
import re 
  
def putSpace(input): 
  
    # regex [A-Z][a-z]* means any string starting  
    # with capital character followed by many  
    # lowercase letters  
    words = re.findall('[A-Z][a-z]*', input) 
  
    # Change first letter of each word into lower 
    # case 
    result = [] 
    for word in words: 
        word = chr( ord (word[0]) + 32) + word[1:] 
        result.append(word) 
    print (' ').join(result) 
```
